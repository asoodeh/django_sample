</div>
<div style="background-color:lightyellow;color:rgb(255,0,255); text-align:left; vertical-align: middle; padding:1px;">
 <h4>website/views.py</h4><hr>
</div>

```python
def post_post(request,pid):
    #your code
    post_object=Post.objects.get(pid=post_id)
```


<div style="background-color:rgba(0, 250, 0, 0.05); text-align:left; vertical-align: middle; padding:10px;">
<a href="https://docs.djangoproject.com/en/5.0/topics/http/shortcuts/"><h2> Model <span style="color:red"> Django </span> shortcut functions</h2></a>
</div>
<ul>
    <li>redirect()</li>
    <li>get_object_or_404()</li>
    <li>get_list_or_404()</li>   
</ul>

```python
from django.shortcuts import render, get_object_or_404, redirect

def my_view(request):
    posts = get_object_or_404(Post, pk=pid)
    context = {'posts': posts}
     return render(request, 'blog/blog-single.html', context)

def my_view(request):
    ...
    return redirect("https://example.com/")     
```

<div style="background-color:rgba(0, 250, 250, 0.05); text-align:left; vertical-align: middle; padding:10px;">
 <b>usin  <span style="color:red"><,>,<=,>= </span> in filter with <span style="color:red">__gte, __gt, __lte, __lt </span> </b> 
</div>
     
```python    
posts = Post.objects.filter(published_date__lte=datetime.now())  # published_date<=datetime.now()
posts = Post.objects.filter(published_date__lt=datetime.now())   # published_date<datetime.now()
posts = Post.objects.filter(published_date__lgte=datetime.now())  # published_date=>datetime.now()
posts = Post.objects.filter(published_date__gt=datetime.now())  # published_date>datetime.now()
```

<div style="background-color:rgba(0, 250, 0, 0.05); text-align:left; vertical-align: middle; padding:10px;">
<a href="https://docs.djangoproject.com/en/5.0/ref/contrib/admin/"><h2> Model <span style="color:red"> Meta </span>options</h2></a>
</div>

<div style="background-color:rgba(0, 250, 0, 0.05); text-align:left; vertical-align: middle; padding:10px;">
<a href="https://www.adamsmith.haus/python/docs/django.db.models.ForeignKey"><h2>  <span style="color:red"> ForeignKey </span></h2></a>
</div>
<div style="background-color:rgba(0, 250, 0, 0.05); text-align:left; vertical-align: middle; padding:10px;">
<a href="https://learndjango.com/tutorials/django-best-practices-referencing-user-model"><h2> Django Best Practices:  <span style="color:red"> Referencing </span> the User Model</h2></a>
</div>

<div style="background-color:rgba(0, 250, 250, 0.05); text-align:left; vertical-align: middle; padding:10px;">
 <h2>three step </h2>create directory <span style="color:red">media </span> and <span style="color:red">website,blog,img,</span> 
</div>

<div style="background-color:rgba(0, 250, 250, 0.05); text-align:left; vertical-align: middle; padding:10px;">
 <h2 style="color:red">Upload Images and ForeignKey</h2>
</div>
<div style="background-color:rgba(0, 250, 0, 0.05); text-align:left; vertical-align: middle; padding:10px;">
<a href="https://docs.djangoproject.com/en/3.2/topics/files/"><h2> Using files in models <span style="color:red"> Managing files </span> </h2></a>
</div>
</div>
<div style="background-color:rgba(0, 250, 0, 0.05); text-align:left; vertical-align: middle; padding:10px;">
<a href="https://www.geeksforgeeks.org/imagefield-django-models/"><h2> <span style="color:red"> ImageField  </span> – Django Models</h2></a>
</div>
</div>
<div style="background-color:rgba(0, 250, 0, 0.05); text-align:left; vertical-align: middle; padding:10px;">
<a href="https://djangocentral.com/uploading-images-with-django/"><h2> How To <span style="color:red"> Upload Images</span> With Django  </h2></a>
</div>
<div style="background-color:lightyellow;color:rgb(255,0,255); text-align:left; vertical-align: middle; padding:1px;">
 <h4>blog/models.py</h4><hr>
</div>

```python
from django.db import models
from django.contrib.auth.models import User

# Create your models here.
class Post(models.Model):
    image = models.ImageField(upload_to='blog/img/', default='blog/img/default.jpg')
    author = models.ForeignKey(User, on_delete=models.SET_NULL, null=True)
    title = models.CharField(max_length=255)
    content = models.TextField()
    # tage
    # category
    counted_views = models.IntegerField(default=0)
    status = models.BooleanField(default=False)
    published_date = models.DateTimeField(null=True)
    created_date = models.DateTimeField(auto_now_add=True)
    updated_date = models.DateTimeField(auto_now=True)
    class Meta:
        ordering = ["-published_date"]
        verbose_name = "پست"
        verbose_name_plural = "پست ها"

    def __str__(self):
       return "%s %s" % (self.title, self.id)


    def increment_counted_views(self):
        self.counted_views += 1
        self.save()

```
<div style="background-color:lightyellow;color:rgb(255,0,255); text-align:left; vertical-align: middle; padding:1px;">
 <h4>blog/admins.py</h4><hr>
</div>

```python
from django.contrib import admin
from blog.models import Post

# Register your models here.
@admin.register(Post)
class PostAdmin(admin.ModelAdmin):
    date_hierarchy = 'created_date'
    empty_value_display = "-empty-"
    fields = ('title', 'image', 'author', 'content', 'counted_views', 'status', 'published_date')    # show fields
    exclude = [""]    # != field
    list_display = ('title', 'image', 'author', 'counted_views', 'status', 'published_date', 'created_date', 'updated_date') # show fields 
    list_filter = ('counted_views', 'author', 'status', 'published_date')
    search_fields = ['title', 'author', 'content', ]
```

<div style="background-color:lightyellow;color:rgb(255,0,255); text-align:left; vertical-align: middle; padding:1px;">
 <h4>website/views.py</h4><hr>
</div>

```python
from django.shortcuts import render, get_object_or_404
from django.http import HttpResponse
from blog.models import Post
from django.utils import timezone

# Create your views here.
def blog_test(request):
    post = Post.objects.all()
    context = {'Firstname': 'ali', 'Lastname': 'asoodeh', 'Email': 'asoodeh.ali@gmail.com', 'post': post}
    return render(request, 'blog/test.html', context)

def blog_view(request):
    posts = Post.objects.filter(published_date__lte=timezone.now())
    context = {'posts': posts}
    return render(request, 'blog/blog-home.html', context)

def blog_single(request, pid):
     posts = Post.objects.filter(status=1)
     post = get_object_or_404(posts, pk=pid)
     post.increment_counted_views()
     context = {'post': post}
     return render(request, 'blog/blog-single.html', context)
```

<div style="background-color:lightyellow;color:rgb(255,0,255); text-align:left; vertical-align: middle; padding:1px;">
 <h4>template/blog/blog_home.html</h4><hr>
</div>

```html
<div class="user-details row">
    <p class="user-name col-lg-12 col-md-12 col-6"><a href="#">{{post.author.get_full_name}}</a> <span class="lnr lnr-user"></span></p>
    <p class="date col-lg-12 col-md-12 col-6"><a href="#">{{post.published_date}}</a> <span class="lnr lnr-calendar-full"></span></p>
    <p class="view col-lg-12 col-md-12 col-6"><a href="#">{{post.counted_views}}</a> <span class="lnr lnr-eye"></span></p>
    <p class="comments col-lg-12 col-md-12 col-6"><a href="#">{{post.status}}</a> <span class="lnr lnr-bubble"></span></p>
</div>
<div class="col-lg-9 col-md-9 ">
    <div class="feature-img"><img class="img-fluid" src="{{post.image.url}}" alt=""> </div>
    <a class="posts-title" href="{% url 'blog:single' pid=post.id %}"><h3>{{post.title}}</h3></a>
    <p class="excert"> {{post.content}}</p>
    <a href="{% url 'blog:single' pid=post.id  %}" class="primary-btn">View More</a>
</div>

```

<div style="background-color:rgba(0, 250, 0, 0.05); text-align:left; vertical-align: middle; padding:10px;">
<a href="https://docs.djangoproject.com/en/3.2/ref/templates/builtins/#date"><h2>  <span style="color:red">  date </span> </h2></a>
</div>
</div>
<div style="background-color:rgba(0, 250, 0, 0.05); text-align:left; vertical-align: middle; padding:10px;">
<a href="https://simpleisbetterthancomplex.com/references/2016/06/21/date-filter.html"><h2> <span style="color:red"></span>  </h2></a>
</div>
</div>
